# What is binary tree

`A binary tree is a tree data structure in which each node can have at most two children, which are referred to as the left child and the right child.`

Overview of binary tree: https://www.geeksforgeeks.org/introduction-to-binary-tree-data-structure-and-algorithm-tutorials/?ref=ghm

Types of binary tree: https://www.geeksforgeeks.org/types-of-binary-tree/

Properties of Tree DS: https://www.javatpoint.com/tree#:~:text=A%20tree%20data%20structure%20is%20a%20non%2Dlinear%20data%20structure,can%20be%20of%20any%20type.

Properties of Binary Tree: https://www.geeksforgeeks.org/properties-of-binary-tree/

`Height of a node:` Number of Edges of the longest path from that node to leaf node.

Example: The height of 'c' node is 2. [c -> m -> h or i]

`Depth of a node:` Lenght of the path from root node to that node.

Example: The Depth of 'c' node is 2. [P -> R -> C]

`Level of a Tree:` Level of tree is the height of tree.

### Implementation of Binary Tree

### Let us create a simple tree with 4 nodes. The created tree would be as follows.



```python
class Node:
    def __init__(self, key):
        self.left = None
        self.right = None
        self.value = key

        
if __name__ == '__main__':
    # create root node
    root = Node(key=3)
    ''' following is the tree after above statement
     3
    / \
    None None'''
    # add left node value and right node value
    root.left = Node(4)
    root.right = Node(5)
    ''' 4 and 5 become left and right children of 3
       3
      /  \
     4    5
    / \  / \
   None None None None'''
    root.left.left = Node(6)
    '''6 becomes left child of 4
        3
       / \
      4   5
     / \ / \
    6 None None None
   / \
   None None'''
```
