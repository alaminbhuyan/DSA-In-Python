# Linear Search

```python
import numpy as np
```

```python
def linear_search(arr, num):
    for i in range(len(arr)):
        if num == arr[i]:
            return i
    return -1
```

```python
arr = np.array([1,2,3,4,5,6,7])

result = linear_search(arr=arr, num=1)

if result == -1:
    print("The number is not found")
else:
    print(f"The number is found at the index position of {result}")
```

# Binary Search

Binary Search is a searching algorithm for finding an element's position in a sorted array.

### Iteration Method

```python
# len(arr) - 1 because our indexing start from zero

def binarySearch(arr, num):
    low_index = 0
    high_index = len(arr) - 1
    
    while low_index <= high_index:
        # take only integer value
        # middle_index = low_index + (high_index - low_index)//2 # overcome overflow but in
        # python don't need to worry about it. Use it in some other programming language: C, C++
        middle_index = (low_index + high_index) // 2
        print(middle_index)
        # for best case scenario
        if num == arr[middle_index]:
            return middle_index
        elif num > arr[middle_index]:
            low_index = middle_index + 1
        else:
            high_index = middle_index - 1
    
    return -1
```

```python
arr = np.arange(1,13)
arr
```

```python
index_val = binarySearch(arr=arr, num=5)

if index_val == -1:
    print("The number is not found")
else:
    print(f"The number is found at the index position of {index_val}")
```

### Recursive Method

```python
def r_binarySearch(arr, key, low, high):
    # Check base case
    if low <= high:
        mid = (low + high) // 2
        # for best case
        if key == arr[mid]:
            return mid
        elif key > arr[mid]:
            return r_binarySearch(arr=arr, key=key, low=mid+1, high=high)
        else:
            return r_binarySearch(arr=arr, key=key, low=low, high=mid-1)
    else:
        return -1
```

```python
arr = np.arange(1,13)

print("array: ", arr)

print("length of array: ", len(arr))

index_val = r_binarySearch(arr=arr, key=7, low=0, high=len(arr)-1)

if index_val == -1:
    print("The number is not found")
else:
    print(f"The number is found at the index position of {index_val}")
```
